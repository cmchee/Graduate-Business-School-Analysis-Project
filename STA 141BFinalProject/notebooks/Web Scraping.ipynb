# Web Scraping

```python
import json
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Scraping the first page of U.S. News website
headers= {"User-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36"}
res = requests.get("https://www.usnews.com/best-graduate-schools/api/search?program=top-business-schools&_page=1", headers=headers)
data = res.json()
info = data["data"]["items"]
df = pd.DataFrame(info)
df.head()
```

```python
# Get the total number of Pages
```

```python
total_page = data["data"]["totalPages"]
total_page
```

```python
#Loop through all the pages and scrape the data from each page
names = []
cities = []
states = []
urls = []
full_time_tuition = []
mlb_ft_enrolled_dir_pg = []
b_test_score = []
rankings = []
for page in range(2, total_page+1):
    print("page", page)
    res = requests.get("https://www.usnews.com/best-graduate-schools/api/search?program=top-business-schools&_page="+str(page),
                       headers=headers)
    data = res.json()
    info = data["data"]["items"]
    temp = pd.DataFrame(info)
    df = pd.concat([df, temp])
```

```python
# Data Cleaning. Gather parameters
```

```python
import re

def get_tuition(info):
    '''
    This function gets the tuition for each program
    '''
    if info["v_ft_tuition"] is None:
        return np.nan
    tuition = info["v_ft_tuition"][0][1]
    tuition= re.findall("\d+,?\d+", tuition)[0]
    tuition = tuition.replace(",", "")
    tuition = float(tuition)
    return tuition

def get_enroll(info):
    '''
    This function gets the enrollment number
    '''
    enroll = info["mlb_ft_enrolled_dir_pg"]
    if enroll is None:
        return np.nan
    enroll = enroll.replace(",", "")
    enroll = int(enroll)
    return enroll

def get_ranking(info):
    '''
    This function gets the rank of school program
    '''
    return info["display_rank"]


df["tuition"] =df["schoolData"].apply(get_tuition)
df["enroll"]=df["schoolData"].apply(get_enroll)
df["rank"]=df["ranking"].apply(get_ranking)
df.head()
```

```python
#clean so it looks readable
schools = df[["name", "city", "state", "url", "rank","tuition","enroll", "rank",  "blurb"]]
schools.head()
```

```python
#Scape website using Beautiful soup to gather the annual salaries and employment rate
salaries = []
employment_rates = []
i = 0
for url in schools.url.values:
    i += 1
    if(i%20==0):
        print(i)
    page = requests.get(url, headers=headers)
    content = page.text
    soup = BeautifulSoup(content)
    divs = soup.find_all("div", {"class":"DataField__Data-qjl95u-2 llIKNt"})
    salary = divs[-2].text.strip()
    employment_rate = divs[-1].text.strip()
    salaries.append(salary)
    employment_rates.append(employment_rate)
```

```python
#make df into csv, also adding employment rate and salaries
schools = schools.copy()
schools["salary"] = salaries
schools["employment_rate"] = employment_rates
```

```python
schools.to_csv("schools.csv")
```
