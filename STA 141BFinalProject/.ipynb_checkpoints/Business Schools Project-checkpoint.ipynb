# Part 2: Comparing Ranked and Unranked Schools

Let us find out if there is a significant difference between the different rank groups. In the dataset, only the top 110 programs are ordered. This group is called the Top 110 schools. The next schools ranked from 110-143 were not ordered by the U.S News, so we will group them as the 'Mid Pack' programs. The last rank group will be the Unranked schools. We will run an ANOVA test and a Tukey's HSD test to find any significant differences between the average salaries. This test was under the null hypothesis that the average salaries between the three rank groups are the same.

```python
#subset table so it's only top 110 schools, edit table so salary and rank are numeric
bus_salary = subset_bus[subset_bus.salary.notnull()]
bus_salary.salary = bus_salary.salary.str.replace('$','')
bus_salary.salary = bus_salary.salary.str.replace(',','')
bus_salary.salary = bus_salary.salary.astype('float')

```

```python
#Top 110 groups vs Mid Pack vs Bottom Half group salaries Anova test
def rank_group1(row):
    '''
    this function is to define groups for the whole subset
    '''
    if row['rank'] == "Unranked":
        return 'Unranked'
    elif row['rank'] == "110-143":
        return 'Mid Pack'
    else:
        return 'Top 110'
    
bus_salary['rank_group'] = bus_salary.apply(lambda row: rank_group1(row), axis=1)
```

```python
print(stats.f_oneway(bus_salary['salary'][bus_salary['rank_group'] == 'Top 110'],
               bus_salary['salary'][bus_salary['rank_group'] == 'Mid Pack'],
               bus_salary['salary'][bus_salary['rank_group'] == 'Unranked'],
              ))
```

```python
#Tukey HSD test to see which groups avg salaries are different, turns out that they are all different from each other.
tukey = pairwise_tukeyhsd(endog=bus_salary['salary'],
                          groups=bus_salary['rank_group'],
                          alpha=0.05)
print(tukey)
```

## Anova and Tukey's HSD Result (Top 110 vs Mid Pack vs Unranked Business Programs)
As we can from the results, we can conclude that there is a difference in average annual salaries between the three rank groups. However, there is no difference in average salaries between the Mid Pack and Unranked programs. 

This tells us that if a person doesn't get accepted to a top ranked school, on average, the person will most likely be making the same amount of money per year as someone who went to an Unranked school. Obviously, there are many other factors such as determination, hard work, luck, and individual differences. So do not be discouraged if you go to an Unranked school because of the average annual salary after graduating. Same goes for being accepted into a top 110 school, later on we will show why all the previous factors do matter as well.

### A Deeper Look at Tuition vs Salary Ranked Schools
If you look at the scatterplot below, notice that when the most of the top schools (dark blue dots) have the highest average salaries after graduating.
```python
#general graph of tuition vs salary
bus_tuition_sal = p9.ggplot(bus_sal_r)
bus_tuition_sal += p9.aes(x = 'tuition', y = 'salary', color = 'rank')
bus_tuition_sal += p9.geom_point() 
bus_tuition_sal += p9.labels.ggtitle("Tuition vs Salary(Ranked Business Schools)")
bus_tuition_sal
```

#### Pearson Correlation of Rank vs Salary
When comparing position rank and average salary, there is an extremely strong correlations with the position rank and salary. This means as the programs are ranked better(lower number), the average salary will be estimated to increase.



```python
#Pearson Correlation of Rank vs Salary
bus_sal_rank_corr = np.corrcoef(bus_sal_r['rank'], bus_sal_r['salary'])
print(bus_sal_rank_corr[0][1])
```

```python
def rank_group(row):
    '''
    This will identify what rank_group each school belongs to. Top 25 means 25-1, top 50 means 50-26,
    Top 75 means  75-51, and Top 110 means 110-76
    '''
    if row['rank'] <= 25:
      return 'Top 25'
    elif row['rank'] <= 50:  
      return 'Top 50'
    elif row['rank'] <=75:
      return 'Top 75'
    elif row['rank'] <= 110:
      return 'Top 110'

    
bus_sal_r['rank_group'] = bus_sal_r.apply(lambda row: rank_group(row), axis=1)
```

### Scatterplot of ranked school salaries vs tuition split by rank group.
Note: 
Top 25 - rank 25-1,
Top 50 - rank 50-26,
Top 75 - rank 75-51,
Top 110 - rank 76-110

```python
#facet_wrap graph of tuition and salary
bus_tuition_sal_wrap = p9.ggplot(bus_sal_r)
bus_tuition_sal_wrap += p9.aes(x = 'tuition', y = 'salary')
bus_tuition_sal_wrap += p9.geom_point() 
bus_tuition_sal_wrap += p9.labels.ggtitle("Tuition vs Salary(Split Ranked Business Schools)")
bus_tuition_sal += p9.facet_wrap('~rank_group')
bus_tuition_sal
```

#### Pearson Correlation of rank vs salary by rank group
The correlations below (Top 25, Top 50, Top 75, Top 110 respectively) indicate the strength correlations between rank and average starting salary start to weaken by a lot.

```python
#pearson correlation of rank and salary per rank_group
print(np.corrcoef(bus_sal_r['rank'][bus_sal_r['rank_group']== 'Top 25'], bus_sal_r['salary'][bus_sal_r['rank_group'] == 'Top 25'])[0][1])
print(np.corrcoef(bus_sal_r['rank'][bus_sal_r['rank_group']== 'Top 50'], bus_sal_r['salary'][bus_sal_r['rank_group'] == 'Top 50'])[0][1])
print(np.corrcoef(bus_sal_r['rank'][bus_sal_r['rank_group']== 'Top 75'], bus_sal_r['salary'][bus_sal_r['rank_group'] == 'Top 75'])[0][1])
print(np.corrcoef(bus_sal_r['rank'][bus_sal_r['rank_group']== 'Top 110'], bus_sal_r['salary'][bus_sal_r['rank_group'] == 'Top 110'])[0][1])
```

### ANOVA and Tukey's HSD Test between the Top 25, Top 50, Top 75, Top 110
The Anova test under the null hypothesis that the average salaries were the same across all ranked groups was rejected at alpha = .05. The Tukey's post-hoc test also indicate that there is a difference in average salaries between all the rank groups. This means that there will usually be a difference in starting salary depending on the rank of your business school.

```python
#anova test with salary in the top 110
print(stats.f_oneway(bus_sal_r['salary'][bus_sal_r['rank_group'] == 'Top 25'],
               bus_sal_r['salary'][bus_sal_r['rank_group'] == 'Top 50'],
               bus_sal_r['salary'][bus_sal_r['rank_group'] == 'Top 75'],
              bus_sal_r['salary'][bus_sal_r['rank_group'] == 'Top 110']))
```

```python
#Tukey HSD test to see which groups avg salaries are different, turns out that they are all different from each other.
tukey = pairwise_tukeyhsd(endog=bus_sal_r['salary'],
                          groups=bus_sal_r['rank_group'],
                          alpha=0.05)
print(tukey)
```

# Part 3 Which Business Programs are the Best?
In part 2, we analyzed the differences in salaries between their rank on U.S. News. Now, we will pick the schools that we think are the best. This analysis will not only include salaries, but also employment rate, enrollment issues, and tuition costs. First, the best schools with average starting salaries of at least 100,000 dollars are subsetted.

```python
#create sql file
sqlite_file = 'schools.sqlite'
sql_engine = sqla.create_engine('sqlite:///' + sqlite_file)

bus_salary.to_sql('business',sql_engine,if_exists='replace')
```

```python
#queries sql to find schools that make at least 100k. This is because I mentioned at the top that being super successful means
#making at least 100k
best_query = """
select * from business
where salary >=100000
"""

best = pd.read_sql_query(best_query,sql_engine)
best['rank'] = best['rank'].astype('int')
best['rank_group'] = best.apply(lambda row: rank_group(row), axis=1)
best
```

```python
#graph of best schools rank vs salary
best.plot(x = 'rank', y = 'salary', title = 'School Rank vs salary')
```

As we can see from the table and the graph, there seems to be a drop in salary the higher the position rank of the program. But let's see which of these high salary schools is the best for its tuition cost!

### Salary to Tuition Ratio
Top schools may result in the highest annual salaries, but their tuitions are usually extremely high as well. Let's see which schools have the highest salary to tuition ratio.

```python
#salary vs tuition ratio, shows only top 10 ratio schools.
best['ratio'] = best['salary'] / best['tuition']
ratio = best.sort_values(by = 'ratio', ascending= False)
ratio10 = ratio.head(10)
ratio10
```

```python
#bar plot of which schools have the best ratio
fig_dims = (11, 6)
fig, ax = plt.subplots(figsize=fig_dims)
ratio_barplot = sns.barplot( x = 'name', y = 'ratio',data = ratio10, palette = "Blues_d", ax= ax)
ratio_barplot.set(xlabel = 'School', ylabel = 'Ratio')
plt.setp(ratio_barplot.xaxis.get_majorticklabels(), rotation=17)
plt.show()
```

Presented above are the top 10 schools with the best Salary to Tuition Ratio. Notice that the top schools with the best ratio only contains one school from the Top 25 group! However, you will also notice that these schools only enroll only a small amount of students into their programs. Therefore, it might be cheap, but it is reasonable to assume that these programs are very tough to get into.

If you were to think about salary alone, obviously the top ranked schools on will most likely result in the highest annual salary.

## You Can't Make Money if You Don't Have a Job

As mentioned before, being able to get into a top 10 school does not guarantee the person to graduate with a job and have a starting salary of 150 thousand dollars per year. It's important to check if the person can even find a job after finishing their business degree.  

```python
#subset data so only top 5 and worst 5 schools
employ = ranked_bus[ranked_bus['employment_rate'].notnull()]
employ['combine'] = employ.name + " " + employ.employment_rate
employ.employment_rate = employ.employment_rate.str.replace('%','')
employ.employment_rate = employ.employment_rate.astype('float')
employ.employment_rate = employ.employment_rate /100
employ = employ[employ.employment_rate != 0]

employ = employ[employ.enroll >=100]
top10low10 = employ.sort_values(by = 'employment_rate', ascending= False)
top10low10 = pd.concat([top10low10.head(5),top10low10.tail(5)]).sort_values(by = 'employment_rate', ascending= True)
```

```python
#create circle radius
circles = circlify.circlify(
    top10low10['employment_rate'].tolist(), 
    show_enclosure=False, 
    target_enclosure=circlify.Circle(x=0, y=0, r=1)
)
```

## Top 5 Best and Worst Schools by Employment Rate (Ranked Schools only with enrollemnt greater than 100 people)
Below is a circle plot of the top 5 schools with the best employment rate and the bottom five schools with the worst employment rate. Because of the lack of data on employment rate for unranked schools, they were not included. Interestingly enough, even if you graduate from Stanford (#1 ranked business school in the nation), it does not show up in the top 5 employment rates. What is also interesting is that North Carolina A&T State University has the highest employment rate at 93.8 percent. The Tuition is also extremely low at around 4500 dollars per year. The average annual salary after graduating is also pretty decent at $77,000. However, the school only enrolls around 150 people. This signals that people looking for jobs should not be overconfident if they are graduating from a top tier business school.  

```python
# Create just a figure and only one subplot
fig, ax = plt.subplots(figsize=(15,15))

# Title
ax.set_title('Employment Rate by School')

# Remove axes
ax.axis('off')

# Find axis boundaries
lim = max(
    max(
        abs(circle.x) + circle.r,
        abs(circle.y) + circle.r,
    )
    for circle in circles
)
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)

# list of labels
labels = top10low10['combine']


# print circles
for circle, label in zip(circles, labels):
    x, y, r = circle
    ax.add_patch(plt.Circle((x, y), r*.9, alpha=.9 ,linewidth=2, facecolor="#69b2a3", edgecolor="black"))
    plt.annotate(label, (x,y ) ,va='center', ha='center', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round', pad=.5))
```
